```python
dados.columns
```

```python
# Função para retirar hifens de uma string

def hyphenator(string):
    
    hyphenless = ''
    
    for char in string:
        if char != '-':
            hyphenless = hyphenless + char
            
    return hyphenless
```

```python
# Função para gerar o dicionario

def dictMaker(df):
    
    cols = df.columns
    dictList = []
    
    for colName in cols:
        if '-' in colName:
            hyphenless = hyphenator(colName)
            dictList.append( (colName,hyphenless) )
            
    dictionary = dict(dictList)
    return dictionary
```

```python
dictMaker(dados)
```

```python
def renameCols(df):
    
    dictionary = dictMaker(df)
    df = df.rename(columns = dictionary)
    
    return df
```

```python
dadosRenomeados = renameCols(dados)
```

```python
dadosRenomeados
```

```python
dadosR = dadosRenomeados
```
